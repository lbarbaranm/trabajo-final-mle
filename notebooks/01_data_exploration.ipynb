```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
data = pd.read_csv('../data/processed/clean_data.csv')

# Visualizar distribuciones de características
plt.hist(data['area'], bins=20)
plt.xlabel('Área')
plt.ylabel('Frecuencia')
plt.title('Distribución del Área')
plt.show()

# Visualizar relaciones entre características
plt.scatter(data['area'], data['price'])
plt.xlabel('Área')
plt.ylabel('Precio')
plt.title('Relación entre Área y Precio')
plt.show()
```
